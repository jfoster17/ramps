```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from astropy.table import Table
from astropy.coordinates import SkyCoord 
from astropy import units as u
```

```python
y = Table.read("pointing-sources.txt",format='ascii')
```

```python
glons = []
glats = []
for i in np.arange(len(y)):
    c = SkyCoord(y['ra'][i] , y['dec'][i] , frame='icrs', unit=(u.hourangle, u.deg))
    glons.append(c.galactic.l.degree)
    glats.append(c.galactic.b.degree)
print(glats)
```

```python
from matplotlib.patches import Rectangle

xs = np.arange(10,50,1)
for x in xs:
    currentAxis = plt.gca()
    currentAxis.add_patch(Rectangle((x - .5, -0.4), 1, 0.8, facecolor="grey",alpha=0.1))


plt.scatter(glons,glats,s=y['flux']*10.)
for i, txt in enumerate(y['name']):
    plt.annotate(txt, (glons[i],glats[i]+1))

plt.xlim(0,50)
plt.ylim(-20,20)
plt.xlabel("Glon [deg]")
plt.ylabel("Glat [deg]")
plt.savefig("K-band-points.pdf")
```
